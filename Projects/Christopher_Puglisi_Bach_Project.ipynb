```python
import pandas as pd

bach1 = pd.read_csv('bach.csv')
bach = bach1.drop('choral_ID', axis=1)
bach
```

```python
bach = pd.concat([pd.get_dummies(bach[['C','C#', 'D',	'D#',	'E',	'F',	'F#',	'G',	'G#',	'A',	'A#',	'B']]), bach[['chord_label']]], axis=1)
bach
```

```python
bFeatures = bach.drop('chord_label', axis=1)
bLabels = bach['chord_label']
from sklearn.model_selection import train_test_split
bach_train_features, bach_test_features, bach_train_labels, bach_test_labels = train_test_split(bFeatures, bLabels, test_size = 0.2, random_state=42)
```

```python
from numpy.random.mtrand import rand
from sklearn import tree
cls = tree.DecisionTreeClassifier(criterion='entropy', max_depth = 13, random_state = 0)
cls.fit(bach_train_features, bach_train_labels)
```

```python
bach_predictions = cls.predict(bach_test_features)
from sklearn.metrics import accuracy_score
accuracy_score(bach_test_labels, bach_predictions)
```
